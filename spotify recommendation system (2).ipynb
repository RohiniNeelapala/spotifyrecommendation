```python
from sklearn.neighbors import NearestNeighbors
```

```python
data=pd.read_csv('spotify.csv')
```

```python
song_name = input('song_name')
recommendations_count = int(input('recommendations_count'))

```

```python
input_song = data[data['song'].str.lower()==song_name.lower()].iloc[0]
input_features =input_song[['acousticness','danceability','energy','instrumentalness','liveness', 'loudness', 'speechiness', 'tempo','valence']]

   
```

```python
 #Fit a k-nearest neighbors model on the dataset
 nbrs=NearestNeighbors(n_neighbors=recommendations_count,metric='euclidean').fit(data[['acousticness','danceability','energy','instrumentalness','liveness', 'loudness', 'speechiness', 'tempo','valence']])

   
```

```python
 #Find the indices of the closest songs to the input song
distances,indices = nbrs.kneighbors(input_features.values.reshape(1, -1))

    
```

```python
#Get the recommendations from the indices
recommendations = data.iloc[indices.flatten()][['artists', 'song']]

#Remove duplicates and convert to list
recommendations = recommendations.drop_duplicates(subset='artists')
```
